## 1. CAC (Customer Acquisition Cost)

### 1.1. Monthly expenses

```python
# Load data
monthly_expense_gs_id = '10OGbaywwMIqKgnPGy8VDvpBVtjyqln47iYa2lFhI9Mw'
monthly_expense_url = f'https://docs.google.com/spreadsheets/d/{monthly_expense_gs_id}/export?format=xlsx'

monthly_expense_df = pd.read_excel(monthly_expense_url, sheet_name='Sheet1')
monthly_expense_df
```

```python
monthly_expense_df.info()
```

```python
# Filter last month data
max_month_expense = monthly_expense_df['month'].max()

last_month_expense_df = monthly_expense_df[monthly_expense_df['month'] == max_month_expense]
last_month_expense_df.head()
```

```python
# Calculate monthly expense cost
crm_cost = last_month_expense_df[last_month_expense_df['item'] == 'Salesforce']['amount'].values[0]
crm_cost
```

### 1.2. Payroll

```python
# Load data
payroll_gs_id = '1c_WihqTZCQvNgxzmd-OwhR9i5diwtfxXVLyMn8R-Lp4'
payroll_url = f'https://docs.google.com/spreadsheets/d/{payroll_gs_id}/export?format=xlsx'

payroll_df = pd.read_excel(payroll_url, sheet_name='Sheet1')
payroll_df
```

```python
# Filter last month data
last_month_payroll_df = payroll_df[payroll_df['month'] == max_month_expense]
last_month_payroll_df.head()
```

```python
# Calculate sales and marketing payroll cost
last_month_sales_marketing_payroll_cost = last_month_payroll_df[last_month_payroll_df['department'].isin(['Sales', 'Marketing'])]['paid'].sum()
last_month_sales_marketing_payroll_cost
```

### 1.3. Marketing spendings

```python
# Load data
marketing_spendings_gs_id = '1AZOIThOV4P-0eYDge53ZwumVkfkHoYPWxst3k3Bv87c'
marketing_spendings_url = f'https://docs.google.com/spreadsheets/d/{marketing_spendings_gs_id}/export?format=xlsx'

marketing_spendings_df = pd.read_excel(marketing_spendings_url, sheet_name='Sheet1')
marketing_spendings_df
```

```python
target_month = max_month_expense.to_period('M')
target_month
```

```python
marketing_spendings_df['date'].dt.to_period('M')
```

```python
# Calculate marketing spending cost
marketing_spendings_cost = marketing_spendings_df[marketing_spendings_df['date'].dt.to_period('M') == target_month]['spending'].sum()
```

```python
# Calculate total sales and marketing expenses
total_sales_marketing_expense = crm_cost + last_month_sales_marketing_payroll_cost + marketing_spendings_cost
total_sales_marketing_expense
```

### 1.4. Number of customers

```python
# Load data
receipt_hs_gs_id = '1qayqML1zCKdmtzutkcy9LWvE6xFRm6TGBEVkHHJKIuE'
receipt_url = f'https://docs.google.com/spreadsheets/d/{receipt_hs_gs_id}/export?format=xlsx'

receipt_df = pd.read_excel(receipt_url, sheet_name='Sheet1')
receipt_df
```

```python
last_month_receipt_df = receipt_df[receipt_df['date'].dt.to_period('M') == target_month]
last_month_receipt_df
```

```python
new_customer_df = last_month_receipt_df[last_month_receipt_df['new_customer'] == 1]
num_new_customer = len(new_customer_df)
num_new_customer
```

### 1.5. Calculate CAC

```python
CAC = total_sales_marketing_expense / num_new_customer
CAC
```

## 2. ARPU (Average Revenue Per User)

### 2.1. Calculate total revenue

```python
total_revenue = last_month_receipt_df['receipt_amount'].sum()
total_revenue
```

### 2.2. Count number of users

```python
num_users = len(last_month_receipt_df['customer_id'].unique())
num_users
```

### 2.3. Calculate ARPU

```python
ARPU = total_revenue / num_users
ARPU
```

## 3. COGS (Cost of Goods Sold)

### 3.1. Calculate production cost

```python
last_month_expense_df
```

```python
last_month_expense_df[last_month_expense_df['category'].isin(['Server Costs','Software Licenses'])]
```

```python
work_items = ['AWS Hosting','Google Cloud Storage','Atlassian Jira']
shared_items = ['Slack', 'Zoom']

#Assume 60% cost for Slack and Zoom is allocated to COGS
work_items_expense = last_month_expense_df[last_month_expense_df['item'].isin(work_items)]['amount'].sum()
shared_items_expense = last_month_expense_df[last_month_expense_df['item'].isin(shared_items)]['amount'].sum()

server_n_software_expense = work_items_expense + shared_items_expense * 0.6
server_n_software_expense
```

### 3.2. Calculate production payroll

```python
production_payroll = last_month_payroll_df[last_month_payroll_df['department'] == 'Engineering']['paid'].sum()
production_payroll
```

### 3.3. Calculate COGS

```python
COGS = server_n_software_expense + production_payroll
COGS
```

## 4. Gross margin

```python
Gross_margin = (total_revenue - COGS) / total_revenue *100
Gross_margin
```

## 5. LTV (Customer Lifetime Value)

```python
# Load data
customer_lifespan_gs_id = '1by8tPHwOnq3uKYK2E7sA9VBUYoPM4p1Rnrm_Ss9cyHI'
customer_lifespan_url = f'https://docs.google.com/spreadsheets/d/{customer_lifespan_gs_id}/export?format=xlsx'

lifespan_df = pd.read_excel(customer_lifespan_url, sheet_name='Sheet1')
lifespan_df
```

```python
lifespan_df['lifespan_days'] = (lifespan_df['churn_date'] - lifespan_df['start_date']).dt.days
lifespan_df
```

```python
# Calculate the average lifespan in days
avg_lifespan_days = lifespan_df['lifespan_days'].mean()

# Convert the average lifespan to months
avg_lifespan_months = avg_lifespan_days / 30
avg_lifespan_months
```

```python
# Calculate LTV
LTV = ARPU * avg_lifespan_months * Gross_margin
LTV
```
